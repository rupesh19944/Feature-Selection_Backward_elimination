```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("smoking.csv")
```

```python
#df['Urine protein'].plot.kde(figsize=(20,10))
#df['serum creatinine'].plot.kde(figsize=(20,10))
df['age'].plot.kde(figsize=(20,10))
```

```python
X = df.iloc[:, 1:-1]
Y = df.iloc[:, -1]
```

```python
pd.set_option('display.max_columns', None)
X.head()
```

```python
X.isnull().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder 
```

```python
one_hot = OneHotEncoder()
```

```python
df1 =one_hot.fit_transform(X[['gender', 'oral','tartar']])
```

```python
df1.toarray()
```

```python
df2 = pd.DataFrame(df1.toarray(), index= X.index)
```

```python
X1 = pd.concat([X, df2], axis=1)
```

```python
X1.drop(['gender', 'oral','tartar'], inplace=True, axis=1)
```

```python
X1.shape
```

```python
X1.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X1, Y, test_size= 0.2, random_state= 0)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
rf= RandomForestClassifier()
rf.fit(X_train, Y_train)
# predict the result
Y_pred = rf.predict(X_test)
score = accuracy_score(Y_test, Y_pred)
print(score)
confusion_matrix(Y_test, Y_pred)
```

Using Rnadom forest feature importance

```python
#getting feature importance
importance = rf.feature_importances_
```

```python
importance
```

```python
importance_df = pd.DataFrame({"Feature": X1.columns, "Importance": importance})
importance_df = importance_df.sort_values('Importance').set_index('Feature')
importance_df
```

```python
importance_df.plot.barh(figsize=(10,15))
```

```python
top_20_imp_feature = importance_df.index[-20:]
```

```python
top_20_imp_feature
```

```python
X_new = X1[top_20_imp_feature]
```

```python
X_new.head()
```

```python
X_train_new, X_test_new, Y_train_new, Y_test_new = train_test_split(X_new, Y, test_size= 0.2, random_state= 0)
```

```python
rf_new= RandomForestClassifier()
rf_new.fit(X_train_new, Y_train_new)
# predict the result
Y_pred_new = rf_new.predict(X_test_new)
score = accuracy_score(Y_test_new, Y_pred_new)
print(score)
confusion_matrix(Y_test_new, Y_pred_new)
```

#Observation 
With top 20 feature predictor can able to predict with same accuracy 

```python
top_10_imp_feature = importance_df.index[-10:]
X_new_10 = X1[top_10_imp_feature]
X_train_new, X_test_new, Y_train_new, Y_test_new = train_test_split(X_new_10, Y, test_size= 0.2, random_state= 0)

rf_new_10= RandomForestClassifier()
rf_new_10.fit(X_train_new, Y_train_new)
# predict the result
Y_pred_new = rf_new_10.predict(X_test_new)
score = accuracy_score(Y_test_new, Y_pred_new)
print(score)
confusion_matrix(Y_test_new, Y_pred_new)
```
