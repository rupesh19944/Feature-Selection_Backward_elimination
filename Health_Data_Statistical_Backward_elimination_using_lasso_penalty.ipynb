```python
X = df.iloc[:, 1:-1]
Y = df.iloc[:, -1]
```

```python
pd.set_option('display.max_columns', None)
X.head()
```

```python
X.isnull().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder 
```

```python
one_hot = OneHotEncoder()
```

```python
df1 =one_hot.fit_transform(X[['gender', 'oral','tartar']])
```

```python
df1.toarray()
```

```python
df2 = pd.DataFrame(df1.toarray(), index= X.index)
```

```python
X1 = pd.concat([X, df2], axis=1)
```

```python
X1.drop(['gender', 'oral','tartar'], inplace=True, axis=1)
```

```python
X1.shape
```

```python
X1.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X1, Y, test_size= 0.2, random_state= 0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
regressor= LogisticRegression()
regressor.fit(X_train, Y_train)
# predict the result
Y_pred = regressor.predict(X_test)
score = accuracy_score(Y_test, Y_pred)
print(score)
confusion_matrix(Y_test, Y_pred)
```

Using Lasso Regularization

```python
log_reg = LogisticRegression(C=0.001,penalty = "l1", solver = 'liblinear', random_state = 0)
log_reg.fit(X1, Y)
# Fitting the model for feature selection
#model = SelectFromModel(log_reg, prefit = True) 
```

```python
log_reg.coef_
```

```python
feature = []
y = 0
for i in (log_reg.coef_[0]):
    if(i != 0):
        feature.append(y)
        y +=1
    else:
        y +=1
feature
```

```python
X_new = X1.iloc[:,feature]
```

```python
X_new.head()
```

```python
X_new.shape
```

```python
X_train_new, X_test_new, Y_train_new, Y_test_new = train_test_split(X_new, Y, test_size= 0.2, random_state= 0)
```

```python
regressor_after= LogisticRegression()
regressor_after.fit(X_train_new, Y_train_new)
# predict the result
Y_pred_new = regressor_after.predict(X_test_new)
score = accuracy_score(Y_test_new, Y_pred_new)
print(score)
confusion_matrix(Y_test_new, Y_pred_new)
```
